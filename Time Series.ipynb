## Time Series - Group Project (Merge datasets)

```python
#Imports
import pandas as pd
pd.options.mode.chained_assignment = None
#Filenames
POPULATION_FILENAME = "Country-Population.xls"
FERTILITYRATE_FILENAME = "Fertility-Rate.xls"
LIFEEXPECTANCY_FILENAME = "Life-Expectancy-At-Birth.xls"
#Import all data
population = pd.read_excel(POPULATION_FILENAME, header = 0)
population.set_index("Country Name", inplace=True)
fertilityRate = pd.read_excel(FERTILITYRATE_FILENAME, header = 0)
fertilityRate.set_index("Country Name", inplace=True)
lifeExpectancy = pd.read_excel(LIFEEXPECTANCY_FILENAME, header = 0)
lifeExpectancy.set_index("Country Name", inplace=True)
```

```python
#Countries we want to visualize
countriesToVisualize = ["United States"]
#Columns these columns
dropColumns = ["Country Code", "Indicator Name", "Indicator Code"]
```

```python
#Grab all data associated with countries (countriesToVisualize)
populationData = population.loc[population.index.isin(countriesToVisualize)]
fertilityRateData = fertilityRate.loc[fertilityRate.index.isin(countriesToVisualize)]
lifeExpectancyData = lifeExpectancy.loc[lifeExpectancy.index.isin(countriesToVisualize)]
```

```python
#Drop unnecessary columns and apply transpose
populationData.drop(dropColumns, axis=1, inplace= True)
populationData.index = "Population_" + populationData.index.astype(str)
populationData = populationData.transpose().rename_axis('Year', axis=1)
populationData = populationData.astype(int)
populationData
```

```python
#Drop unnecessary columns and apply transpose
fertilityRateData.drop(dropColumns, axis=1, inplace= True)
fertilityRateData.index = "FertilityRate_" + fertilityRateData.index.astype(str)
fertilityRateData = fertilityRateData.transpose().rename_axis('Year', axis=1)
fertilityRateData
```

```python
#Drop unnecessary columns and apply transpose
lifeExpectancyData.drop(dropColumns, axis=1, inplace= True)
lifeExpectancyData.index = "LifeExpectancy_" + lifeExpectancyData.index.astype(str)
lifeExpectancyData = lifeExpectancyData.transpose().rename_axis('Year', axis=1)
lifeExpectancyData
```

```python
#Merge all dataset
finalDataset = populationData.merge(fertilityRateData, left_index=True, right_index=True, how='inner').merge(lifeExpectancyData, left_index=True, right_index=True, how='inner')
finalDataset
```

```python
#Re-order year column and create a CSV
finalDataset.insert(loc=0, column="Year", value = finalDataset.index)
finalDataset.to_csv("TimeSeries.csv", index = False);
```
